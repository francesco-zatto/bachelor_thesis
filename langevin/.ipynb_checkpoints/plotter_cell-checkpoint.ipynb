```python
import matplotlib.pyplot as plt
import pandas as pd
from random import randint
```

```python
df = pd.read_csv("data.csv", sep=";").iloc[:, :-1]
```

```python
for i in range(0, len(df.columns), 2):
    plot = plt.plot(df.iloc[:, i], df.iloc[:, i + 1]) #every position
    plt.plot(df.iloc[0, i], df.iloc[0, i + 1], c="black", marker="*") #start
    plt.plot(df.iloc[-1, i], df.iloc[-1, i + 1], c=plot[0].get_color(), marker="*") #end
plt.savefig("images/plot_discrete001.png")
```
