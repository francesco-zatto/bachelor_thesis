```python
import matplotlib.pyplot as plt
import pandas as pd
from random import randint
```

```python
df = pd.read_csv("data.csv", sep=";").iloc[:, :-1]
```

```python
for i in range(0, len(df.columns), 2):
    plt.plot(df.iloc[:, i], df.iloc[:, i + 1]) #every position
    plt.plot(df.iloc[0, i], df.iloc[0, i + 1], c="red", marker="*") #start
    plt.plot(df.iloc[-1, i], df.iloc[-1, i + 1], c="red", marker="*") #end
#plt.savefig("images/plot" + str(randint(0, 10_000)) + ".png")
```
